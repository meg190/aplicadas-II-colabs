##Contexto

* Una empresa de transporte urbano opera $4$ rutas ($R_1$, $R_2$, $R_3$ y $R_4$). Cada ruta incurre mensualmente en costos por:

  - Conductor: $C$ (costos por turno de conductor).
  - Autobús: $A$ (costos por bus asignado; corresponde a arrendamiento/amortización).
  - Combustible: $F$ (cantidad de combustible consumido).
  - Mantenimiento: $M$ (costo mantenimiento programada).

* Para el mes analizado, la planificación operativa (unidades consumidas por ruta) y los costos totales mensuales por ruta son:

  1. En la Ruta 1 ($R_{1}$) se emplean 3 conductores, 3 autobuses, 5 hectolitros  de combustible y 1 revisión de mantenimiento. El costo total es 13770 euros.
  2. En la Ruta 2 ($R_{2}$) se utilizan 5 conductores, 3 autobuses, 3 hectolitros de combustible y 4 revisiones de mantenimiento. El costo total es 17512 euros.
  3. En la Ruta 3 ($R_{3}$) se necesitan 3 conductores, 3 autobuses, 5 hectolitros de combustible y 3 revisiones de mantenimiento. El costo total es 14270 euros.
  4. En la Ruta 4 ($R_{4}$) se usan 4 conductores, 2 autobuses, 4 hectolitros de combustible y 0 revisiones de mantenimiento. El costo total es 12016 euros.

### Preguntas

Empezamos planteando el problema en forma de matriz:

$$\left[\begin{array}{cccc|c}
3 & 3 & 5 & 1  & 13770 \\

\end{array}\right]
$$

Para resolver un sistema matricial utilizamos el método de Gauss-Jordan, el cual consiste en aplicar una serie de operaciones elementales para transformar la parte izquierda de la matriz aumentada en la matriz de identidad.

La matriz identidad es aquella en la que los elementos de la diagonal principal (de izquierda a derecha) son unos, y todos los demás elementos son ceros.

Por ejemplo, si deseamos que el primer elemento de una fila pase de tener el valor 3 a 1, debemos multiplicar toda la fila por el inverso multiplicativo de 3, que es ( \frac{1}{3} ). De esta forma, obtenemos una nueva fila en la que el valor correspondiente se convierte en 1, acercándonos al objetivo de formar la matriz identidad.


$$F_1 \cdot \frac{1}{3} \rightarrow F_1$$
$$
\left[\begin{array}{cccc|c}
1 & 1 & \frac{5}{3} & \frac{1}{3} & 4590  \\
5 & 3 & 3 & 4 & 17512\\
3 & 3 & 5 & 3 & 14270\\
4 & 2 & 4 & 0 & 12016
\end{array}\right]
$$

Una vez que hemos obtenido un 1 en la posición deseada (llamado pivote), el siguiente paso es hacer que todos los demás valores en esa misma columna se conviertan en 0.

Para lograrlo, tomamos la fila donde se encuentra el pivote y la multiplicamos por el valor que deseamos eliminar, pero con signo contrario. Luego, sumamos el resultado a la fila que contiene ese valor.

Este procedimiento se repite con cada fila de la matriz, excepto con la fila del pivote. Así, conseguimos que en la columna del pivote solo quede un 1 y los demás elementos sean 0, avanzando hacia la forma reducida por filas de Gauss-Jordan.


$$F_1 \cdot -5  + F_2 \rightarrow F_2$$
$$
\left[\begin{array}{cccc|c}

0 & -2 & -\frac{16}{3} & \frac{7}{3} & -5438\\
3 & 3 & 5 & 3 & 14270\\
4 & 2 & 4 & 0 & 12016
\end{array}\right]
$$
$$F_1 ⋅ -3 + F_3 \rightarrow F_3$$
$$
\left[\begin{array}{cccc|c}

$$$$

$$$$
\left[\begin{array}{cccc|c}

0 & -2 & -\frac{8}{3} & -\frac{4}{3} & -6344
\end{array}\right]
$$
Intercambiamos la fila 3 con la fila 4 porque en la posición donde necesitamos colocar un 1 actualmente hay un 0.

Este paso es necesario ya que el pivote el elemento que debe ser igual a 1 para continuar con el proceso de eliminación no puede ser cero. Al realizar el intercambio, ubicamos una fila con un valor distinto de cero en la posición adecuada, lo que nos permite continuar con el método de Gauss-Jordan sin interrupciones.

En el siguiente paso, repetimos el mismo procedimiento: transformamos el siguiente valor de la diagonal principal en 1 y convertimos todos los demás valores de su columna en 0.

Este proceso se realiza sucesivamente para cada fila, avanzando de arriba hacia abajo, hasta que toda la parte izquierda de la matriz se transforme completamente en una matriz identidad. De esta manera, habremos resuelto el sistema por el método de Gauss-Jordan.

1. $I1$ influye en $I2$(0.8), $I4$(0.6), $I5$(0.7)

##Análisis de influencia global

Para entender la influencia en el grafo, se debe aclarar que es un grafo dirigido, es decir que está influenciado por pesos. En un grafo dirigido, cada conexión tiene una dirección (quién influye sobre quién) y un peso (qué tan fuerte es esa influencia).Por lo tanto:.

*   Un peso saliente representa cuánto influye un nodo sobre otros
*   Un peso entrante representa cuánto es influenciado un nodo por los demás.

###1. ¿Qué influencer tiene la mayor suma de pesos salientes (es decir, influye más en los demás)?

Teniendo en cuenta el contexto anterior, para poder responder esta pregunta, se deben sumar todos los pesos de las flechas que salen de cada influencer (Pesos salientes).Estas sumas se realizan mediante el siguiente codigo:

```python
influencias = {'I1': {'I2': 0.8, 'I4': 0.6, 'I5': 0.7},'I2': {'I3': 0.9, 'I4': 0.5},'I3': {'I1': 0.4, 'I4': 0.8},'I4': {'I5': 0.9},'I5': {'I2': 0.6, 'I3': 0.7}}
salientes = {}

for emisor, conexiones in influencias.items():
    total = sum(conexiones.values())  # sumamos los pesos que emite
    salientes[emisor] = total

print("Pesos salientes por influencer (quién influye más en los demás):")
for influencer, total in salientes.items():
    print(f"{influencer}: {round(total, 1)}")

mas_influyente = max(salientes, key=salientes.get)

print(f"\nEl influencer que más influye en los demás es: {mas_influyente} "
      f"con un total de {round(salientes[mas_influyente], 1)}")
```

Segun el calculo anterior, I1 tiene mayor suma de pesos salientes, es decir que es el más influyente, ya que si vemos la matriz, tiene poder directo sobre tres nodos: I2, I4 e I5. Por lo tanto, I1 es el centro emisor de la matriz, es decir, el nodo que más envía influencia porque es el de mayores pesos salientes.

###2. ¿Qué influencer tiene la mayor suma de pesos entrantes (es decir, es más influenciado por otros)?
Siguiendo el mismo razonamiento, para saber quien es el más influenciado se deben sumar todos los pesos de las flechas que apuntan hacia cada influencer (Pesos entrantes).

```python
influencias = {'I1': {'I2': 0.8, 'I4': 0.6, 'I5': 0.7},'I2': {'I3': 0.9, 'I4': 0.5},'I3': {'I1': 0.4, 'I4': 0.8},'I4': {'I5': 0.9},'I5': {'I2': 0.6, 'I3': 0.7}}

entrantes = {'I1': 0, 'I2': 0, 'I3': 0, 'I4': 0, 'I5': 0}

for emisor, conexiones in influencias.items():
    for receptor, peso in conexiones.items():
        entrantes[receptor] += peso
print("Pesos entrantes por influencer:")
for influencer, total in entrantes.items():
    print(f"{influencer}: {round(total, 1)}")

mas_influenciado = max(entrantes, key=entrantes.get)

print(f"\nEl influencer más influenciado por otros es: {mas_influenciado} con un total de {round(entrantes[mas_influenciado], 1)}")
```

Según el resultado del código anterior, I4 es el más influenciable ya que recibe influencia desde muchos nodos (I1, I2, e I3), por lo tanto, I4 actúa como punto de convergencia de la información en la matriz.

El influencer I1 es el punto de origen de la campaña: inicialmente, solo él promociona el producto.
A partir de ahí, los demás influencers pueden adoptarlo con una probabilidad proporcional a la influencia que reciben de I1 o de quienes ya lo hayan adoptado. Partiendo de lo anterior, I1 ya tiene el producto porque su probabilidad es de 1 y la de los demás que no tienen el producto es de 0.
En una red de influencias, no todos adoptan al mismo tiempo.
Cada influencer necesita recibir suficiente influencia acumulada antes de decidir adoptar.
Por eso, el proceso se divide en pasos discretos (Paso 1, Paso 2, etc.), que representan etapas sucesivas de propagación.

| I4         | 0.6                       | 1.9               | 0.32                     |
| I5         | 0.7                       | 1.6               | 0.44                     |

Viendo ahora la nueva influencia, I2 es el más propenso a adoptar el producto en esta etapa (57 %), seguido por I5 y I4. Por el contrario, I3 aún no tiene influencia directa, por lo que no adopta en este paso.

* Paso 2: Influecias secundarias

Ahora I2, I4, I5 adoptaron el producto, por lo que influencian a sus respectivos conectados: I2 influye a I3 e I4, I4 influye a I5, I5 influye en I2 y I3. Como resultado, en este paso todos los influencers lo adoptan. En este caso aunque en el paso 1 I3 no estuvo influenciado y en el paso dos fue el ultimo en adoptar el producto,fue e, que más se vio afectado por la influencia de I2 e I5 por lo que la probabilidad acumulada de adopción para I3 se estimaría alrededor del 45 % al final del segundo paso.
