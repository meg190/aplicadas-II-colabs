  3. En la Ruta 3 ($R_{3}$) se necesitan 3 conductores, 3 autobuses, 5 hectolitros de combustible y 3 revisiones de mantenimiento. El costo total es 14270 euros.

Para resolver un sistema matricial utilizamos el m√©todo de Gauss-Jordan, el cual consiste en aplicar una serie de operaciones elementales para transformar la parte izquierda de la matriz aumentada en la matriz de identidad.

La matriz identidad es aquella en la que los elementos de la diagonal principal (de izquierda a derecha) son unos, y todos los dem√°s elementos son ceros.

Por ejemplo, si deseamos que el primer elemento de una fila pase de tener el valor 3 a 1, debemos multiplicar toda la fila por el inverso multiplicativo de 3, que es ( \frac{1}{3} ). De esta forma, obtenemos una nueva fila en la que el valor correspondiente se convierte en 1, acerc√°ndonos al objetivo de formar la matriz identidad.


$$F_1 \cdot \frac{1}{3} \rightarrow F_1$$
$$
\left[\begin{array}{cccc|c}
1 & 1 & \frac{5}{3} & \frac{1}{3} & 4590  \\
5 & 3 & 3 & 4 & 17512\\
3 & 3 & 5 & 3 & 14270\\
4 & 2 & 4 & 0 & 12016
\end{array}\right]
$$

Una vez que hemos obtenido un 1 en la posici√≥n deseada (llamado pivote), el siguiente paso es hacer que todos los dem√°s valores en esa misma columna se conviertan en 0.

Para lograrlo, tomamos la fila donde se encuentra el pivote y la multiplicamos por el valor que deseamos eliminar, pero con signo contrario. Luego, sumamos el resultado a la fila que contiene ese valor.

Este procedimiento se repite con cada fila de la matriz, excepto con la fila del pivote. As√≠, conseguimos que en la columna del pivote solo quede un 1 y los dem√°s elementos sean 0, avanzando hacia la forma reducida por filas de Gauss-Jordan.

Al realizar el intercambio, ubicamos una fila con un valor distinto de cero en la posici√≥n adecuada, lo que nos permite continuar con el m√©todo de Gauss-Jordan sin interrupciones.

Este proceso se realiza sucesivamente para cada fila, avanzando de arriba hacia abajo, hasta que toda la parte izquierda de la matriz se transforme completamente en una matriz identidad. De esta manera, habremos resuelto el sistema por el m√©todo de Gauss-Jordan.

##An√°lisis de influencia global

Para entender la influencia en el grafo, se debe aclarar que es un grafo dirigido, es decir que est√° influenciado por pesos. En un grafo dirigido, cada conexi√≥n tiene una direcci√≥n (qui√©n influye sobre qui√©n) y un peso (qu√© tan fuerte es esa influencia).Por lo tanto:.

*   Un peso saliente representa cu√°nto influye un nodo sobre otros
*   Un peso entrante representa cu√°nto es influenciado un nodo por los dem√°s.

###1. ¬øQu√© influencer tiene la mayor suma de pesos salientes (es decir, influye m√°s en los dem√°s)?

Teniendo en cuenta el contexto anterior, para poder responder esta pregunta, se deben sumar todos los pesos de las flechas que salen de cada influencer (Pesos salientes).Estas sumas se realizan mediante el siguiente codigo:

```python
influencias = {'I1': {'I2': 0.8, 'I4': 0.6, 'I5': 0.7},'I2': {'I3': 0.9, 'I4': 0.5},'I3': {'I1': 0.4, 'I4': 0.8},'I4': {'I5': 0.9},'I5': {'I2': 0.6, 'I3': 0.7}}
salientes = {}

for emisor, conexiones in influencias.items():
    total = sum(conexiones.values())  # sumamos los pesos que emite
    salientes[emisor] = total

print("Pesos salientes por influencer (qui√©n influye m√°s en los dem√°s):")
for influencer, total in salientes.items():
    print(f"{influencer}: {round(total, 1)}")

mas_influyente = max(salientes, key=salientes.get)

print(f"\nEl influencer que m√°s influye en los dem√°s es: {mas_influyente} "
      f"con un total de {round(salientes[mas_influyente], 1)}")
```

Segun el calculo anterior, I1 tiene mayor suma de pesos salientes, es decir que es el m√°s influyente, ya que si vemos la matriz, tiene poder directo sobre tres nodos: I2, I4 e I5. Por lo tanto, I1 es el centro emisor de la matriz, es decir, el nodo que m√°s env√≠a influencia porque es el de mayores pesos salientes.

###2. ¬øQu√© influencer tiene la mayor suma de pesos entrantes (es decir, es m√°s influenciado por otros)?
Siguiendo el mismo razonamiento, para saber quien es el m√°s influenciado se deben sumar todos los pesos de las flechas que apuntan hacia cada influencer (Pesos entrantes).

```python
influencias = {'I1': {'I2': 0.8, 'I4': 0.6, 'I5': 0.7},'I2': {'I3': 0.9, 'I4': 0.5},'I3': {'I1': 0.4, 'I4': 0.8},'I4': {'I5': 0.9},'I5': {'I2': 0.6, 'I3': 0.7}}

entrantes = {'I1': 0, 'I2': 0, 'I3': 0, 'I4': 0, 'I5': 0}

for emisor, conexiones in influencias.items():
    for receptor, peso in conexiones.items():
        entrantes[receptor] += peso
print("Pesos entrantes por influencer:")
for influencer, total in entrantes.items():
    print(f"{influencer}: {round(total, 1)}")

mas_influenciado = max(entrantes, key=entrantes.get)

print(f"\nEl influencer m√°s influenciado por otros es: {mas_influenciado} con un total de {round(entrantes[mas_influenciado], 1)}")
```

Seg√∫n el resultado del c√≥digo anterior, I4 es el m√°s influenciable ya que recibe influencia desde muchos nodos (I1, I2, e I3), por lo tanto, I4 act√∫a como punto de convergencia de la informaci√≥n en la matriz.
